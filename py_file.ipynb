```python
median_salary_year_avg = df['salary_year_avg'].median()
median_salary_hour_avg = df['salary_hour_avg'].median()
```

```python
new_df = df
new_df['salary_year_avg'] = new_df['salary_year_avg'].fillna(median_salary_year_avg)
new_df['salary_hour_avg'] = new_df['salary_hour_avg'].fillna(median_salary_hour_avg)
new_df.loc[0:, 'salary_year_avg':'salary_hour_avg']
```

```python
df_unic = new_df
df_unic.drop_duplicates()
```

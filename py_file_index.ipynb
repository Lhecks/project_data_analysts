```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('csv/job_postings_fact.csv')
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_posted_month'] = df.job_posted_date.dt.month
df.index.name = 'index_job'
df
```

```python
cnd_job = df[df['job_country']=='Canada']
cnd_job
```

```python
cnd_job.reset_index(inplace=True)
cnd_job
```

```python
cnd_job.set_index('index_job', inplace=True)
cnd_job
```

```python
cnd_j = cnd_job.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['mean','min','max','median'])
cnd_j
```

```python
cnd_j.sort_values(by=('median','salary_year_avg'), ascending=False)
```

```python
cnd_j.plot(kind='bar')
plt.xticks(rotation=45, ha='right')
plt.show()
```

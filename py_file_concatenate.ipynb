##### Quarter Job's Trend
In this project, we are going to plot the number of pe term and also to the jo title
Each term will have two bar chart the first will be for the number of jobs and the second will be for the top 3 job title

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('csv/job_postings_fact.csv')
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_posted_month'] = df.job_posted_date.dt.strftime('%b')
df
```

```python
df_months = df['job_posted_month'].unique()
# df_months
{months : months for months in df_months}
months_dict = {months: df[df['job_posted_month'] == months] for months in df_months}
months_dict['Dec']
```

```python
conc_q1_dt = pd.concat([months_dict['Jan'], months_dict['Feb'], months_dict['Mar']], ignore_index=True)
conc_q1_dt['job_posted_month'].value_counts().plot(kind='bar')
plt.xlabel('Number of job Q1')
plt.xticks(rotation=0)
plt.ylim(0, 100000)
plt.show()
```

```python
# Let's get first the top 3 job titles for Q1
top_job_q1 = conc_q1_dt['job_title_short'].value_counts().head(3)
top_job_list_q1 = top_job_q1.index.to_list()

# Filtering the DataFrame so that to get the top job titles
filtered_df_q1 = conc_q1_dt[conc_q1_dt['job_title_short'].isin(top_job_list_q1)].copy()

# Let's define the order of the months of the first Quater Q1
month_order_q1 = ['Jan', 'Feb', 'Mar']

# Convert the 'job_posted_month' column to a categorical type with the specified order
filtered_df_q1['job_posted_month'] = pd.Categorical(filtered_df_q1['job_posted_month'], categories=month_order_q1, ordered=True)

grouped_data = filtered_df_q1.groupby(['job_posted_month', 'job_title_short'], observed=False).size().unstack(fill_value=0)

# Plot the grouped data
grouped_data.plot(kind='bar', stacked=False)
plt.xlabel('Q1')
plt.ylabel('Number of Jobs')
plt.title('Top 3 Job Titles per Month in Q1')
plt.xticks(rotation=0)
plt.legend(title='Job Title')
plt.ylim(0, 30000)
plt.show()
```

```python
# Let's define the order of the months of the first Quater Q2
month_order_q2 = ['Apr', 'May', 'Jun']

# Concatenate the data for Q2
conc_q2_dt = pd.concat([months_dict['Apr'], months_dict['May'], months_dict['Jun']], ignore_index=True)

# Convert the 'job_posted_month' column to a categorical type with the specified order
conc_q2_dt['job_posted_month'] = pd.Categorical(conc_q2_dt['job_posted_month'], categories=month_order_q2, ordered=True)

# Count the occurrences of each month and plot them
month_counts = conc_q2_dt['job_posted_month'].value_counts().sort_index()

# Plot the data
month_counts.plot(kind='bar')
plt.xticks(rotation=0)
plt.xlabel('Month')
plt.ylabel('Number of Jobs')
plt.title('Number of Jobs in Q2')
plt.ylim(0, 100000)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
```

```python
df = pd.read_csv('csv/job_postings_fact.csv')
skill_job = pd.read_csv('csv/skills_job_dim.csv')
skill_df = pd.read_csv('csv/skills_dim.csv')
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_posted_month'] = df.job_posted_date.dt.strftime('%b')
# df.index.name = 'index_job'
df
```

```python
fisrt_merged = df.merge(skill_job, on='job_id')
fisrt_merged
```

```python
second_merge = fisrt_merged.merge(skill_df, on='skill_id')
inflation_df = second_merge[pd.notna(second_merge['salary_year_avg'])].copy()
second_merge
```

```python
second_merge = second_merge.drop(columns='skill_id')
```

```python
second_merge.info()
```

```python
def projection_wages_inflation(salary):
    return salary * 1.5
```

```python
inflation_df['inflation_salary'] = inflation_df['salary_hour_avg'].apply(projection_wages_inflation)
inflation_df[['salary_year_avg', 'inflation_salary']]
```

```python
cnd_job = second_merge[second_merge['job_country']=='Canada'].copy()
cnd_job_pivot = cnd_job.pivot_table(columns='job_title_short', index='job_posted_month', aggfunc='size')
cnd_job_pivot.reset_index(inplace=True)
cnd_job_pivot['month_number'] = pd.to_datetime(cnd_job_pivot['job_posted_month'], format='%b').dt.month
cnd_job_pivot.sort_values(by='month_number', inplace=True)
cnd_job_pivot.set_index('job_posted_month', inplace=True)
cnd_job_pivot.drop(columns='month_number', inplace=True)
cnd_job_pivot
```

```python
top_job = cnd_job['job_title_short'].value_counts().head(3)
top_job = top_job.index.to_list()
cnd_job_pivot[top_job].plot(kind='line')
plt.title('The Top 3 Monthly Job in Canada')
plt.xlabel('2023')
plt.ylabel('Number of Job')
plt.show()
```

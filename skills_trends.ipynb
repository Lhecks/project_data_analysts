```python
import pandas as pd
import matplotlib.pyplot as plt 
df = pd.read_csv('csv/new_dataset.csv')
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_posted_month'] = df.job_posted_date.dt.strftime('%b')
df['job_posted_month_number'] = df['job_posted_date'].dt.month
df = df.drop(columns=df.columns[0])
df.set_index('job_id')
```

```python
# df = df['job_posted_month_number'] = df['job_posted_date'].dt.month.apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
```

```python
df_pivot = df[df['job_title_short'] == 'Data Analyst'].copy()
df_pivot
```

```python
df_pivot = df_pivot.pivot_table(index='job_posted_month_number', columns='skills', aggfunc='size', fill_value=0)
df_pivot
# print(df['salary_year_avg'].isna().sum())
```

```python
df_pivot.loc['Total'] = df_pivot.sum()
df_pivot = df_pivot[df_pivot.loc['Total'].sort_values(ascending=False).index]
df_p = df_pivot.drop('Total')
```

```python
df_p.iloc[:,:5].plot(
    kind='line',
    linewidth=2,
    linestyle='--',
    marker='*',
    markersize=8,
    colormap='inferno',
    figsize=(10,5)
    )
plt.title('Top Skills Trends Over the Year of Data Analyst')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```

```python
skills_count = df.groupby(['skills','job_title_short']).size()
skills_count
```

```python
df_skills = skills_count.reset_index(name='skills_count')
df_skills.sort_values(by='skills_count', ascending=False, inplace=True)
```

```python
df_skills
```

```python
job_title_list = ['Data Analyst','Data Scientist','Data Engineer']
```

```python
fig, ax =plt.subplots(3,1)
for i, job_title in enumerate(job_title_list):
    # print(i, job_title)
    df_plot = df_skills[df_skills['job_title_short'] == job_title].head()
    df_plot.plot(kind='barh', ax=ax[i], title=job_title, figsize=(10,8))
    # print(df_plot)
    
    plt.xlabel('Job title')
    ax[i].invert_yaxis()
    ax[i].set_ylabel('')
    ax[i].set_xlim(0, 120000)
    ax[i].legend().set_visible(False)

fig.suptitle('Skills Comparison Between Data Analyst - Scientist - Engineer', fontsize=12)
fig.tight_layout()
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', startangle=90, autopct='%1.2f%%')
plt.ylabel('')
plt.title('Chart to ')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1,3)
fig.set_size_inches(15,5)
dict_items = {
    'job_work_from_home':'Work From Home',
    'job_no_degree_mention':'No Degree Required',
    'job_health_insurance': 'Health Insurance'
}

for i, (desc, title) in enumerate(dict_items.items()):
    # print(f'{i}-->i,  {desc}: Desc, {title}: title')
    ax[i].pie(df[desc].value_counts(), startangle=90, autopct='%1.2f%%', labels=["False", 'True'])
    ax[i].set_title(title)
plt.show()
```

```python

fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# Dictionary to map column names to titles
dict_items = {
    'job_work_from_home': 'Work From Home',
    'job_no_degree_mention': 'No Degree Required',
    'job_health_insurance': 'Health Insurance'
}

colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = (0.1, 0)  # explode 1st slice for emphasis

for i, (desc, title) in enumerate(dict_items.items()):
    values = df[desc].value_counts()
    labels = ["False", "True"]
    
    wedges, texts, autotexts = ax[i].pie(
        values,
        startangle=90,
        autopct='%1.2f%%',
        labels=labels,
        colors=colors,
        explode=explode,
        shadow=True
    )

    ax[i].set_title(title, fontsize=16, fontweight='bold')
    
    # Customizing font properties
    for text in texts + autotexts:
        text.set_fontsize(12)
    
    autotexts[0].set_color('white')
    autotexts[1].set_color('white')
    
    # Adding a legend
    ax[i].legend(wedges, labels, title="Status", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Adjust layout to make room for titles and legends
plt.tight_layout()
plt.show()
```

```python
df_ex = df[df['job_title_short'] == 'Data Analyst']
# df_ex
skills_statistics = df_ex.groupby('skills').agg(
    skills_counts = ('skills', 'count'),
    median_salary = ('salary_year_avg', 'median')
)
skills_statistics = skills_statistics.sort_values(by='skills_counts', ascending=False).head(10)
```

```python
# fig, ax = plt.subplots()
import matplotlib.ticker as tck
skills_statistics.plot(kind='scatter', x='skills_counts', y='median_salary', figsize=(10,6))
plt.xlabel('Median Yearly Salary')
plt.ylabel('Skills Count')
plt.title('Top 5 Skills for Job Posting - Salary ', fontsize=20)
plt.tight_layout
for i, text in enumerate(skills_statistics.index):
    plt.text(skills_statistics['skills_counts'].iloc[i], skills_statistics['median_salary'].iloc[i], text)

ax = plt.gca()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, pos:f'${int(y/1000)}K'))
ax.xaxis.set_major_formatter(tck.FuncFormatter(lambda x, pos:f'{int(x):,}'))

plt.show()
```

```python
df_cnd = df[(df['job_title_short']==job_title_list[0]) & (df['job_country'] == 'Canada')].copy()
df_cnd = df_cnd.dropna(subset=['salary_year_avg'])
df_us = df[(df['job_title_short']==job_title_list[0]) & (df['job_country'] == 'United States')].copy()
df_us = df_us.dropna(subset=['salary_year_avg'])
```

```python
df_cnd['salary_year_avg'].plot(kind='hist', bins=20, edgecolor='black', figsize=(10, 6))
# plt.xlim(0, 125000)
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos:f'${int(x/1000)}K'))
plt.title('Canada Distribution of the yearly salary in the Data Analyst')
plt.xlabel('Salary Per Year')

plt.ylabel('Number of Jobs')
plt.show()
```

```python
df_us['salary_year_avg'].plot(kind='hist', bins=50, edgecolor='black', figsize=(10, 6))
plt.xlim(0, 250000)
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos:f'${int(x/1000)}K'))
plt.title('USA Distribution of the yearly salary in the Data Analyst')
plt.xlabel('Salary Per Year')

plt.ylabel('Number of Jobs')
plt.show()
```

```python
df_us['salary_year_avg'].plot(kind='box', vert=False, figsize=(10, 6))
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos:f'${int(x/1000)}K'))
plt.show()
```

```python
# United States
df_can = df[(df['job_title_short'].isin(job_title_list)) & (df['job_country'] == 'Canada')].copy()
df_can = df_can.dropna(subset=['salary_year_avg'])
job_lists  = [df_can[df_can['job_title_short']==job_title]['salary_year_avg'] for job_title in job_title_list]
```

```python
plt.boxplot(job_lists, labels=job_title_list, vert=False)
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos:f'${int(x/1000)}K'))
plt.title('Distribution of Salary in Canada')
plt.xlabel('Salary Per Year ($USD)')
plt.ylabel('')
plt.show()
```

```python
# United States
df_usa = df[(df['job_title_short'].isin(job_title_list)) & (df['job_country'] == 'United States')].copy()
df_usa = df_usa.dropna(subset=['salary_year_avg'])
job_lists_us  = [df_usa[df_usa['job_title_short']==job_title]['salary_year_avg'] for job_title in job_title_list]

plt.boxplot(job_lists_us, labels=job_title_list, vert=False)
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos:f'${int(x/1000)}K'))
plt.title('Distribution of Salary in The United States')
plt.xlabel('Salary Per Year ($USD)')
plt.ylabel('')
plt.show()
```
